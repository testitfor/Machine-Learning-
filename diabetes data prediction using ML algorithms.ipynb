# Data Splitting

```python
x=dia.drop('Outcome',axis=1)
y=dia['Outcome']
```

# standardizing the variables

```python
feature_scale = [feature for feature in dia.columns if feature!='Outcome']
## Importing library 
##This estimator scales and translates each feature individually such that it is in the given range on the training set, e.g. between zero and one.

from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
scaler.fit(x)
```

```python
dia = pd.concat([dia['Outcome'].reset_index(drop=True),
                    pd.DataFrame(scaler.transform(x), columns=feature_scale)],
                    axis=1)
```

```python
dia.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=567)
```

```python
x_train.shape,y_train.shape
```

```python
x_test.shape,y_test.shape
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import ConfusionMatrixDisplay,accuracy_score,confusion_matrix
from sklearn import tree
```

# LogisticRegression

```python
accuracy=[]
model = LogisticRegression()
model.fit(x_train,y_train)
y_pred = model.predict(x_test)
ConfusionMatrixDisplay.from_predictions(y_test,y_pred)
accuracy.append(accuracy_score(y_test,y_pred))
```

```python
y_pred
```

```python
accuracy
```

# DecisionTreeClassifier

```python
accuracy=[]
model = DecisionTreeClassifier()
model.fit(x_train,y_train)
y_pred = model.predict(x_test)
ConfusionMatrixDisplay.from_predictions(y_test,y_pred)
accuracy.append(accuracy_score(y_test,y_pred))
```

# KNeighborsClassifier

```python
accuracy=[]
model = KNeighborsClassifier()
model.fit(x_train,y_train)
y_pred = model.predict(x_test)
ConfusionMatrixDisplay.from_predictions(y_test,y_pred)
accuracy.append(accuracy_score(y_test,y_pred))
```

```python
y_pred
```

```python
accuracy
```

```python
scores=[]
for i in range(1,32):
    classifier=KNeighborsClassifier(n_neighbors=i).fit(x_train,y_train)
    y_pred=classifier.predict(x_test)
    acc_score=accuracy_score(y_test, y_pred)
    scores.append(accuracy_score(y_pred,y_test))
        
print(scores)   
```

```python
plt.plot(range(1,32),scores)
plt.xlabel("values of k")
plt.ylabel("accuracy")
```
