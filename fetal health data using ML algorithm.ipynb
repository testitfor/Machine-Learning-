```python
data=data[(data.mean_value_of_short_term_variability<10)]
data=data[(data.percentage_of_time_with_abnormal_long_term_variability<20)]
data=data[(data.histogram_max<200)]
data=data[(data.histogram_mode<165) & (data.histogram_mode>100)]
data=data[(data.histogram_mode<165) & (data.histogram_mode>100)]
data=data[(data.histogram_mean<165) & (data.histogram_mean>100)]
data=data[(data.histogram_median<165) & (data.histogram_median>100)]
data=data[(data.histogram_variance<40)]
data=data[(data.mean_value_of_long_term_variability<25)]
data=data[(data.percentage_of_time_with_abnormal_long_term_variability<10)]


data.plot(kind="box",subplots=True,layout=(5,5),sharex=False,sharey=True,figsize=(20,15))

```

```python
x=data.drop('fetal_health',axis=1)
y=data['fetal_health']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=567)
```

```python
x_train.shape,y_train.shape
```

```python
x_test.shape,y_test.shape
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import ConfusionMatrixDisplay,accuracy_score,confusion_matrix
from sklearn import tree
```

# LogisticRegression

```python
accuracy=[]
model = LogisticRegression()
model.fit(x_train,y_train)
y_pred = model.predict(x_test)
ConfusionMatrixDisplay.from_predictions(y_test,y_pred)
accuracy.append(accuracy_score(y_test,y_pred))
```

```python
y_pred
```

```python
accuracy
```

# DecisionTreeClassifier

```python
accuracy=[]
model = DecisionTreeClassifier()
model.fit(x_train,y_train)
y_pred = model.predict(x_test)
ConfusionMatrixDisplay.from_predictions(y_test,y_pred)
accuracy.append(accuracy_score(y_test,y_pred))
```

```python
y_pred
```

# KNeighborsClassifier

```python
accuracy=[]
model = KNeighborsClassifier()
model.fit(x_train,y_train)
y_pred = model.predict(x_test)
ConfusionMatrixDisplay.from_predictions(y_test,y_pred)
accuracy.append(accuracy_score(y_test,y_pred))
```

```python
y_pred
```

```python
scores=[]
for i in range(1,32):
    classifier=KNeighborsClassifier(n_neighbors=i).fit(x_train,y_train)
    y_pred=classifier.predict(x_test)
    acc_score=accuracy_score(y_test, y_pred)
    scores.append(accuracy_score(y_pred,y_test))
        
print(scores)   
```

```python
plt.plot(range(1,32),scores)
plt.xlabel("values of k")
plt.ylabel("accuracy")
```
